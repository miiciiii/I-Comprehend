```python
import os
import numpy as np
```

```python
BASE_DIR = os.getcwd()

IMAGE_DATA_PATH = os.path.join(BASE_DIR, '..', 'datasets', 'processed', 'image_data.npy')

image_data = np.load(IMAGE_DATA_PATH)
print(f"Image Data Shape: {image_data.shape}")
# Expected Output: (number_of_samples, height, width, channels)
```

```python
LABELS_DATA_PATH = os.path.join(BASE_DIR, '..', 'datasets', 'processed', 'image_labels_encoded.npy')

labels_data = np.load(LABELS_DATA_PATH)
print(f"Labels Data Shape: {labels_data.shape}")
# Expected Output: (number_of_samples,)
```

```python
assert image_data.shape[0] == labels_data.shape[0], "Mismatch between number of images and labels"
```

```python
unique_labels = np.unique(labels_data)
print("Unique labels:", unique_labels)
```

```python
# Print first few labels
print("First 10 labels:", labels_data[:10])
```

```python
import matplotlib.pyplot as plt

def display_samples(images, labels, num_samples=5):
    plt.figure(figsize=(15, 5))
    for i in range(num_samples):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(images[i], cmap='gray')  # Adjust cmap if not grayscale
        plt.title(f"Label: {labels[i]}")
        plt.axis('off')
    plt.show()

# Assuming image_data is already loaded
display_samples(image_data, labels_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming labels_data contains class indices (e.g., 0, 1, 2, ...)
unique, counts = np.unique(labels_data, return_counts=True)
plt.bar(unique, counts)
plt.xlabel('Class Labels')
plt.ylabel('Frequency')
plt.title('Class Distribution')
plt.show()

# Print counts for each class
for label, count in zip(unique, counts):
    print(f"Class {label}: {count} samples")
```
